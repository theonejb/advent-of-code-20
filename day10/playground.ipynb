```python
# Solution from https://www.youtube.com/watch?v=6Jnf362ee-E&feature=youtu.be
# I could not figure this one out :(
def solve(adapters):
    paths = [1] + [0] * (len(adapters) - 1)
    for i, adapter in enumerate(adapters):
        for j in range(i - 3, i):
            if(adapter - adapters[j] <= 3):
                paths[i] += paths[j]

    return paths[-1]
```

```python
with open('input.txt') as f:
    problem_input = [0] + sorted(list(
        map(int, f.readlines())
    ))
    print(problem_input)
    print(solve(problem_input))
```
